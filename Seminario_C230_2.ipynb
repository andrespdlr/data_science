```python
from random import randint
import numpy as np
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
def factorial(n):
    if n < 1:
        return 1
    else:
        return n * factorial(n - 1)
```

```python
X = []
Y = []
Z = []
for i in range(20):
    x = randint(0, 100)
    y = randint(0, 100)
    z = 'A-' + str(randint(0, 100))
    X.append(x)
    Y.append(y)
    Z.append(z)
```

```python
from matplotlib.patches import Rectangle
```

```python
plt.figure(figsize=(10, 10))
xs_ticks = np.arange(len(Z))
plt.xticks(xs_ticks, Z)
plt.bar(xs_ticks, Y, color = 'blue', align = 'center', edgecolor = 'black')
```

```python
plt.figure(figsize=(10, 5))
plt.tick_params(axis='both', which='major', labelsize=15)
plt.title("Valores Aleatorios", fontsize = 20, fontweight = 'bold')
plt.xlabel("\nValor X", fontsize = 15)
plt.ylabel("Valor Y\n", fontsize = 15)
plt.ylim(0, 100)
xs_ticks = np.arange(len(X))
plt.xticks(xs_ticks, X)
plt.bar(xs_ticks, Y, color = 'blue', align = 'center', edgecolor = 'black')
plt.savefig('chart_belisario_1.png', dpi = 100, bbox_inches="tight")
```

```python
mean_X = sum(X) / len(X)
mean_Y = sum(Y) / len(Y)
plt.figure(figsize=(12, 12))
axis = plt.gca()
plt.tick_params(axis='both', which='major', labelsize=15)
plt.title("Valores Aleatorios", fontsize = 20, fontweight = 'bold')
plt.xlabel("Valor X", fontsize = 15)
plt.ylabel("Valor Y", fontsize = 15)
plt.xlim(0, 100)
plt.ylim(0, 100)
plt.axvline(mean_X, color = 'black', linestyle = 'solid', linewidth = 2)
plt.axhline(mean_Y, color = 'black', linestyle = 'solid', linewidth = 2)
axis.add_patch(Rectangle((25, 29), 32, 100-mean_Y, facecolor = 'blue', alpha = .25))
plt.plot(X, Y, 'o', c = 'black', markersize = 20, alpha = .10)
marker = np.asarray(Z)
plt.plot(75, 75, '^', c = 'black', markersize = 20)
plt.text(80, 80, 'Andres', color = 'white', ha='left', va = 'top', bbox=dict(facecolor = 'blue', alpha = .25), fontsize = 20)

for i, x in enumerate(X):
    y = Y[i] 
    z = Z[i]
    plt.text(x, y, z, color = 'black', ha='center', va = 'top', fontsize = 10)
    
plt.savefig('chart_belisario_2.png', dpi = 100, bbox_inches="tight", bbox_extra_artists=[])
```
